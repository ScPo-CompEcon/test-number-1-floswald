```julia
using Plots
using LaTeXStrings
q(p) = 2*p^-0.5
p(q) = (q/2)^(-2)
plot(q,0,6,ylim=[0;5],ylabel="q(p)",xlabel="p",legend=false)  
plot!(q,0.97,4.03,fill=(0,0.5,:red))  
plot!([0;p(q(1))],[q(1);q(1)],line=(:black,:dash))
plot!([0;p(q(4))],[q(4);q(4)],line=(:black,:dash))
plot!([1;1],[0;q(1)],line=(:black,:dash))
plot!([4;4],[0;q(4)],line=(:black,:dash))
```

# Analytical Solution

We have for consumer surplus

$$ CS_{p} = \int_p^\infty q(x) dx  $$

and the quantity of interest is then

\begin{align}
CS_4 - CS_1  &= \int_4^\infty q(x) dx - \int_1^\infty q(x) dx \\
             &= 2 \int_1^4 q(x) dx\\
             &= 2 \int_1^4 x^{-0.5} dx \\
             &= 4
\end{align}
